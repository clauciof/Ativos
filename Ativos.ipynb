```python
df = pd.read_csv("1min-17-04 - 1min-17-04.csv")
```

```python
#formatando 'datetime'
df['datetime'] = df['datetime'].map(lambda x: datetime.datetime.fromtimestamp(x/1000))
```

```python
#bradesco
BD = df.loc[df['asset'] == 'BBDC3']
BD.drop(['asset', 'volume_a', 'volume_v', 'volume_i' ], axis=1, inplace=True)

#ambev
AB = df.loc[df['asset'] == 'ABEV3']
AB.drop(['asset', 'volume_a', 'volume_v', 'volume_i' ], axis=1, inplace=True)

#azul
AZ = df.loc[df['asset'] == 'AZUL4']
AZ.drop(['asset', 'volume_a', 'volume_v', 'volume_i' ], axis=1, inplace=True)
```

```python
# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=BD['datetime'], y=BD['low'],
                    mode='lines',
                    name='low '))
fig.add_trace(go.Scatter(x=BD['datetime'],  y=BD['high'],
                    mode='lines',
                    name='high'))


#open/close
fig.add_trace(go.Scatter(x=BD['datetime'], y=BD['open'],
                    mode='lines',
                    name='open '))
fig.add_trace(go.Scatter(x=BD['datetime'],  y=BD['close'],
                    mode='lines',
                    name='close'))

fig.update_layout(title='BBDC4')
fig.show()
```

```python
fig = go.Figure(data=[go.Candlestick(x=BD['datetime'],
                open=BD['open'], high=BD['high'],
                low=BD['low'], close=BD['close'])
                     ])

fig.update_layout(title='BBDC4 Candlestick',xaxis_rangeslider_visible=False)
fig.show()
```
